# Introducción

El CURP, conocido como la Clave Única de Registro de Población, es un código alfanumérico exclusivo empleado en México con el propósito de identificar de manera única a cada individuo. Esta clave desempeña un papel fundamental en una amplia gama de procedimientos y servicios gubernamentales, así como en la identificación de ciudadanos en bases de datos oficiales. El CURP se basa en una serie de directrices particulares que integran elementos esenciales de la información personal de la persona.

##Código en límpio

```python
def obtener_primera_vocal(cadena):
    for letra in cadena:
        if letra in "AEIOU":
            return letra
    return ""

def obtener_entidad_de_nacimiento(entidad):
    entidades = {
        "Aguascalientes": "AS",
        "Baja California": "BC",
        "Baja California Sur": "BS",
        "Campeche": "CC",
        "Coahuila de Zaragoza": "CL",
        "Colima": "CM",
        "Chiapas": "CS",
        "Chihuahua": "CH",
        "Distrito Federal": "DF",
        "Durango": "DG",
        "Gunajuato": "GT",
        "Guerrero": "GR",
        "Hidalgo": "HG",
        "Jalisco": "JC",
        "Edo. de México": "MC",
        "Michoacán de Ocampo": "MN",
        "Morelos": "MS",
        "Nayarit": "NT",
        "Nuevo León": "NL",
        "Oaxaca": "OC",
        "Puebla": "PL",
        "Querétaro": "QT",
        "Quintana Roo": "QR",
        "San Luis Potosí": "SP",
        "Sinaloa": "SL",
        "Sonora": "SR",
        "Tabasco": "TC",
        "Tamaulipas": "TS",
        "Tlaxcala": "TL",
        "Veracruz": "VZ",
        "Yucatán": "YN",
        "Zacatecas": "ZS",
        "Extranjero": "NE",
    }
    return entidades.get(entidad, "OTRO")

def generar_curp(nombre, apellido_paterno, apellido_materno, anio_nacimiento, mes_nacimiento, dia_nacimiento, genero, entidad_federativa):
    primera_letra_apellido = apellido_paterno[0]
    primer_vocal_apellido = obtener_primera_vocal(apellido_paterno)
    primera_letra_segundo_apellido = apellido_materno[0]
    primera_letra_nombre = nombre[0]
    entidad_federativa = obtener_entidad_de_nacimiento(entidad_federativa)
    primera_consonante_apellido = next((letra for letra in apellido_paterno[1:] if letra not in "AEIOU"), "")
    primera_consonante_segundo_apellido = next((letra for letra in apellido_materno[1:] if letra not in "AEIOU"), "")
    primera_consonante_nombre = next((letra for letra in nombre[1:] if letra not in "AEIOU"), "")

    try:
        anio_nacimiento = str(int(anio_nacimiento[-2:]))
        mes_nacimiento = str(int(mes_nacimiento)).zfill(2)
        dia_nacimiento = str(int(dia_nacimiento)).zfill(2)
    except (ValueError, IndexError):
        print("Error: Asegúrate de ingresar valores numéricos válidos para el año, mes y día de nacimiento.")
        return

    if genero not in ("H", "M"):
        print("Error: Ingresa 'H' para Hombre o 'M' para Mujer como género.")
        return

    curp = (
        primera_letra_apellido +
        primer_vocal_apellido +
        primera_letra_segundo_apellido +
        primera_letra_nombre +
        anio_nacimiento +
        mes_nacimiento +
        dia_nacimiento +
        genero +
        entidad_federativa +
        primera_consonante_apellido +
        primera_consonante_segundo_apellido +
        primera_consonante_nombre
    )

    return curp

def main():
    print("¡Bienvenido al generador de CURP!\n Por favor no use tilde en los registros \n")

    while True:
        nombre = input("Ingresa tu primer o único nombre: ").upper()
        if not nombre.isalpha():
            print("Error: El nombre no debe contener números ni caracteres especiales.")
            continue
        else:
            break

    while True:
        apellido_paterno = input("\n Ingresa tu primer apellido: ").upper()
        if not apellido_paterno.isalpha():
            print("Error: El apellido paterno no debe contener números ni caracteres especiales.")
            continue
        else:
            break

    while True:
        apellido_materno = input("\n Ingresa tu segundo apellido: ").upper()
        if not apellido_materno.isalpha():
            print("Error: El apellido materno no debe contener números ni caracteres especiales.")
            continue
        else:
            break

    while True:
        anio_nacimiento = input("\n Ingresa el año de nacimiento (por ejemplo, 2000): ")
        if not anio_nacimiento.isdigit() or not (2 <= len(anio_nacimiento) <= 4):
            print("Error: Asegúrate de ingresar un valor numérico de 2 a 4 dígitos para el año de nacimiento.")
            continue
        else:
            break

    while True:
        mes_nacimiento = input("\n Ingresa el mes de nacimiento (por ejemplo, Abril = 04): ")
        dia_nacimiento = input("\n Ingresa el día de nacimiento (por ejemplo, 22): ")

        if len(mes_nacimiento) != 2 or not mes_nacimiento.isdigit() or len(dia_nacimiento) != 2 or not dia_nacimiento.isdigit():
            print("Error: Asegúrate de ingresar valores numéricos de dos dígitos para el mes y el día de nacimiento.")
            continue
        else:
            break

    while True:
        genero = input("\n Ingresa tu género (H para Hombre, M para Mujer, respeta las mayusculas): ").upper()
        if genero not in ("H", "M"):
            print("Error: Ingresa 'H' para Hombre o 'M' para Mujer como género.")
            continue
        else:
            break

    entidad_federativa = input("\n Ingresa tu entidad federativa de nacimiento: ").title()

    curp_generado = generar_curp(nombre, apellido_paterno, apellido_materno, anio_nacimiento, mes_nacimiento, dia_nacimiento, genero, entidad_federativa)
    if curp_generado:
        print("\n Tu CURP es:", curp_generado)

if __name__ == "__main__":
    main()
```

##Explicación

```python
# Función para obtener la primera vocal de una cadena
def obtener_primera_vocal(cadena): #obtener_primera_vocal, toma una cadena cadenacomo entrada
    for letra in cadena:
        if letra in "AEIOU": #Si encuentra una letra que es una vocal (A, E, I, O o U), la devuelve.
            return letra #Si no se encuentra ninguna voz, devuelve una cadena vacía
    return ""

# Función para obtener la entidad federativa
def obtener_entidad_de_nacimiento(entidad):
    entidades = { #definimos un diccionario con nombre entidades para relacionarlas con sus abreviaturas
        "Aguascalientes": "AS",
        "Baja California": "BC",
        "Baja California Sur": "BS",
        "Campeche": "CC",
        "Coahuila de Zaragoza": "CL",
        "Colima": "CM",
        "Chiapas": "CS",
        "Chihuahua": "CH",
        "Distrito Federal": "DF",
        "Durango": "DG",
        "Gunajuato": "GT",
        "Guerrero": "GR",
        "Hidalgo": "HG",
        "Jalisco": "JC",
        "Edo. de México": "MC",
        "Michoacán de Ocampo": "MN",
        "Morelos": "MS",
        "Nayarit": "NT",
        "Nuevo León": "NL",
        "Oaxaca": "OC",
        "Puebla": "PL",
        "Querétaro": "QT",
        "Quintana Roo": "QR",
        "San Luis Potosí": "SP",
        "Sinaloa": "SL",
        "Sonora": "SR",
        "Tabasco": "TC",
        "Tamaulipas": "TS",
        "Tlaxcala": "TL",
        "Veracruz": "VZ",
        "Yucatán": "YN",
        "Zacatecas": "ZS",
        "Extranjero": "NE",
    }
    return entidades.get(entidad, "OTRO") #Usamos el metodo get para buscar la abreviatura correspondiente. sino se encuentra devuelve "OTRO"

# Función para generar el CURP
def generar_curp(nombre, apellido_paterno, apellido_materno, anio_nacimiento, mes_nacimiento, dia_nacimiento, genero, entidad_federativa):
  #generar_curptoma múltiples argumentos para
    # Obtener las letras correspondientes al CURP
    primera_letra_apellido = apellido_paterno[0]
    primer_vocal_apellido = obtener_primera_vocal(apellido_paterno)
    primera_letra_segundo_apellido = apellido_materno[0]
    primera_letra_nombre = nombre[0]
    entidad_federativa = obtener_entidad_de_nacimiento(entidad_federativa)
    primera_consonante_apellido = next((letra for letra in apellido_paterno[1:] if letra not in "AEIOU"), "")
    primera_consonante_segundo_apellido = next((letra for letra in apellido_materno[1:] if letra not in "AEIOU"), "")
    primera_consonante_nombre = next((letra for letra in nombre[1:] if letra not in "AEIOU"), "")

    # Manejo de excepciones para verificar que el año, mes y día de nacimiento sean números válidos
    # Utilizamos los bloques try y except para capturar cualquier error que pueda ocurrir durante la conversión de cadenas a enteros y formateo de los números.
    try:
        anio_nacimiento = str(int(anio_nacimiento[-2:]))
        mes_nacimiento = str(int(mes_nacimiento)).zfill(2)
        dia_nacimiento = str(int(dia_nacimiento)).zfill(2)
    except (ValueError, IndexError):
        print("Error: Asegúrate de ingresar valores numéricos válidos para el año, mes y día de nacimiento.")
        return

    # Verificar que el género sea válido (H o M)
    if genero not in ("H", "M"):
        print("Error: Ingresa 'H' para Hombre o 'M' para Mujer como género.")
        return

    # Generamos el CURP combianando todas las letras obtenidas anteriormente
    curp = (
        primera_letra_apellido +
        primer_vocal_apellido +
        primera_letra_segundo_apellido +
        primera_letra_nombre +
        anio_nacimiento +
        mes_nacimiento +
        dia_nacimiento +
        genero +
        entidad_federativa +
        primera_consonante_apellido +
        primera_consonante_segundo_apellido +
        primera_consonante_nombre
    )

    return curp

# Función principal
def main(): # función principal del programa que coordina la entrada de datos y la generación de la CURP.
    print("¡Bienvenido al generador de CURP!\n Por favor no use tilde en los registros \n")

#Usamos un bucle para segurarnos que los datos son validos.
    while True:
        nombre = input("Ingresa tu primer o único nombre: ").upper()
        if not nombre.isalpha():
            print("Error: El nombre no debe contener números ni caracteres especiales.")
            continue
        else:
            break

#solicitamos y verificamos los apellidos que solo contenga letras y no contenga números ni caracteres especiales.
    while True:
        apellido_paterno = input("\n Ingresa tu primer apellido: ").upper()
        if not apellido_paterno.isalpha():
            print("Error: El apellido paterno no debe contener números ni caracteres especiales.")
            continue
        else:
            break

    while True:
        apellido_materno = input("\n Ingresa tu segundo apellido: ").upper()
        if not apellido_materno.isalpha():
            print("Error: El apellido materno no debe contener números ni caracteres especiales.")
            continue
        else:
            break

#Aquí se solicita el año de nacimiento. El código verifica que la entrada consta de dígitos y que la longitud de la cadena esté entre 2 y 4 dígitos
    while True:
        anio_nacimiento = input("\n Ingresa el año de nacimiento (por ejemplo, 2000): ")
        if not anio_nacimiento.isdigit() or not (2 <= len(anio_nacimiento) <= 4):
            print("Error: Asegúrate de ingresar un valor numérico de 2 a 4 dígitos para el año de nacimiento.")
            continue
        else:
            break

    # Se implementa un bucle para solicitar el mes y el día de nacimiento, y manejar errores de forma adecuada.
    while True:
        mes_nacimiento = input("\n Ingresa el mes de nacimiento (por ejemplo, Abril = 04): ")
        dia_nacimiento = input("\n Ingresa el día de nacimiento (por ejemplo, 22): ")

        if len(mes_nacimiento) != 2 or not mes_nacimiento.isdigit() or len(dia_nacimiento) != 2 or not dia_nacimiento.isdigit():
            print("Error: Asegúrate de ingresar valores numéricos de dos dígitos para el mes y el día de nacimiento.")
            continue
        else:
            break

#verificamos que sea "H" para Hombre o "M" para Mujer, respetando las mayúsculas.
    while True:
        genero = input("\n Ingresa tu género (H para Hombre, M para Mujer, respeta las mayusculas): ").upper()
        if genero not in ("H", "M"):
            print("Error: Ingresa 'H' para Hombre o 'M' para Mujer como género.")
            continue
        else:
            break

#solicitamos la entidad federativa de nacimiento y se genera la CURP llamando a la función generar_curpcon todos los datos ingresados.
    entidad_federativa = input("\n Ingresa tu entidad federativa de nacimiento: ").title()

#Si se genera una CURP válida, se muestra en la pantalla.
    curp_generado = generar_curp(nombre, apellido_paterno, apellido_materno, anio_nacimiento, mes_nacimiento, dia_nacimiento, genero, entidad_federativa)
    if curp_generado:
        print("\n Tu CURP es:", curp_generado)

if __name__ == "__main__": #se utiliza para determinar si un archivo de Python se ejecuta como un programa independiente o se importa como un módulo en otro programa.
    main() #llamamos a la función para correr el cóodigo
```
