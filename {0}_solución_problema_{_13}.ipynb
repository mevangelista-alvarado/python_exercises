*Impresin de la ecuacin de la recta: Finalmente, se imprime la ecuacin de la recta en funcin de m y b. Dependiendo del valor de b, se formatea la impresin para incluir el signo "+" o "-" antes de b para que la ecuacin sea legible. La ecuacin se imprime con dos decimales:*

```python
def ecuacion_de_recta(punto1, punto2):
    x1, y1 = punto1
    x2, y2 = punto2


    if x1 == x2:
        print("La pendiente es infinita, la recta es vertical.")
        return

    m = (y2 - y1) / (x2 - x1)

    # Calcular la ordenada al origen (b)
    b = y1 - m * x1

    # Imprimir la ecuacin de la recta
    if b < 0:
        print(f"La ecuacin de la recta es: y = {m:.2f}x - {abs(b):.2f}")
    else:
        print(f"La ecuacin de la recta es: y = {m:.2f}x + {b:.2f}")
```
